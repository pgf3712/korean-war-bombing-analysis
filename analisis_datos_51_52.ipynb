# üìä **Descripci√≥n de las columnas del DataFrame**

## **1. Informaci√≥n b√°sica**
1. **`ROW_NUMBER`**: üî¢ N√∫mero de fila o registro √∫nico.
2. **`MISSION_NUMBER`**: üõ©Ô∏è N√∫mero de misi√≥n asignado.
3. **`OP_ORDER`**: üìã Orden operacional asociada a la misi√≥n.
4. **`UNIT`**: üõ°Ô∏è Unidad militar encargada de la misi√≥n.
5. **`AIRCRAFT_TYPE_MDS`**: ‚úàÔ∏è Tipo o modelo del avi√≥n utilizado en la misi√≥n.

---

## **2. Detalles de aeronaves**
6. **`NBR_ATTACK_EFFEC_AIRCRAFT`**: ‚úÖ N√∫mero de aeronaves efectivas en el ataque.
7. **`SORTIE_DUPE`**: üîÑ Duplicado de salidas (para validar datos).
8. **`NBR_ABORT_AIRCRAFT`**: ‚ùå N√∫mero de aeronaves que abortaron la misi√≥n.
9. **`NBR_LOST_AIRCRAFT`**: üí• N√∫mero de aeronaves perdidas durante la misi√≥n.

---

## **3. Informaci√≥n del objetivo**
10. **`TARGET_NAME`**: üéØ Nombre del objetivo.
11. **`TGT_TYPE`**: üè≠ Tipo de objetivo (por ejemplo, infraestructura militar, ciudad, etc.).
12. **`SOURCE_UTM_JAPAN_B`**: üìç Coordenadas UTM del origen del objetivo en Jap√≥n.
13. **`SOURCE_TGT_UTM`**: üìç Coordenadas UTM del origen del objetivo en otro sistema.
14. **`TGT_MGRS`**: üìå Referencia del objetivo en sistema MGRS.
15. **`TGT_LATITUDE_WGS84`**: üåç Latitud del objetivo (sistema WGS84).
16. **`TGT_LONGITUDE_WGS84`**: üåç Longitud del objetivo (sistema WGS84).
17. **`SOURCE_TGT_LAT`**: üó∫Ô∏è Latitud de la fuente de lanzamiento.
18. **`SOURCE_TGT_LONG`**: üó∫Ô∏è Longitud de la fuente de lanzamiento.

---

## **4. Armamento**
19. **`NBR_OF_WEAPONS`**: üî´ N√∫mero de armas utilizadas en la misi√≥n.
20. **`WEAPONS_TYPE`**: üí£ Tipo de armas o municiones empleadas.
21. **`BOMB_SIGHTING_METHOD`**: üëÄ M√©todo de avistamiento para bombardeo.
22. **`TOTAL_BOMBLOAD_IN_LBS`**: ‚öñÔ∏è Carga total de bombas en libras.

---

## **5. Detalles de la misi√≥n**
23. **`TOT`**: üïí Tiempo total de la misi√≥n.
24. **`MISSION_TYPE`**: ‚úàÔ∏è Tipo de misi√≥n (por ejemplo, ataque, reconocimiento, etc.).
25. **`ALTITUDE_FT`**: üìè Altitud de vuelo en pies.
26. **`CALLSIGN`**: üìû Indicativo de la misi√≥n.
27. **`BDA`**: üì∏ Evaluaci√≥n de da√±os en batalla (Battle Damage Assessment).
28. **`NOSE_FUZE`**: üî© Tipo de fusible de la nariz de la bomba.
29. **`TAIL_FUZE`**: üî© Tipo de fusible de la cola de la bomba.
30. **`CALCULATED_BOMBLOAD_LBS`**: ‚öñÔ∏è Carga calculada de bombas en libras.

---

## **6. Informaci√≥n adicional**
31. **`RECORD_SOURCE`**: üìú Fuente de registro de la misi√≥n.
32. **`DAY`**: üìÖ D√≠a de la misi√≥n.
33. **`MONTH`**: üìÖ Mes de la misi√≥n.
34. **`YEAR`**: üìÖ A√±o de la misi√≥n.
35. **`MISSION_DATE`**: üìÖ Fecha completa de la misi√≥n.
36. **`City/Town`**: üèôÔ∏è Ciudad o localidad relacionada con la misi√≥n.
37. **`geometry`**: üåç Representaci√≥n geogr√°fica del objetivo.
38. **`In_Land`**: üèûÔ∏è Indicador de si el objetivo est√° en tierra firme o en el agua.

# LECTURA BASES DE DATOS

```python
file_path = './data/df_bombardeos_procesado.csv'  
df_bombardeos_51_52 = pd.read_csv(file_path, sep=',')                                                                                                                                                                                   
df_bombardeos_51_52.head()
```

```python
# N√∫mero total de misiones
total_missions = df_bombardeos_51_52['MISSION_NUMBER'].nunique()
print(f"Total de misiones: {total_missions}")

# Misiones por tipo
missions_by_type = df_bombardeos_51_52['MISSION_TYPE'].value_counts()
print("\nMisiones por tipo:")
print(missions_by_type)

# Duraci√≥n promedio de las misiones
# Primero, convertimos la columna 'TOT' a un tipo num√©rico, ignorando errores
df_bombardeos_51_52['TOT'] = pd.to_numeric(df_bombardeos_51_52['TOT'], errors='coerce')
average_duration = df_bombardeos_51_52['TOT'].mean()
print(f"\nDuraci√≥n promedio de las misiones: {average_duration:.2f}")

# Altitud promedio de las misiones por tipo
# Primero, convertimos la columna 'ALTITUDE_FT' a un tipo num√©rico, ignorando errores
df_bombardeos_51_52['ALTITUDE_FT'] = pd.to_numeric(df_bombardeos_51_52['ALTITUDE_FT'], errors='coerce')
average_altitude_by_type = df_bombardeos_51_52.groupby('MISSION_TYPE')['ALTITUDE_FT'].mean()
print("\nAltitud promedio de las misiones por tipo:")
print(average_altitude_by_type)
```

```python
# N√∫mero total de aeronaves enviadas
total_effective_aircraft = df_bombardeos_51_52['NBR_ATTACK_EFFEC_AIRCRAFT'].sum()
print(f"N√∫mero total de aeronaves efectivas: {total_effective_aircraft}")

# Tasa de abortos de misi√≥n
# Primero, eliminamos los valores NaN para evitar errores en la divisi√≥n
df_bombardeos_51_52['NBR_ABORT_AIRCRAFT'] = df_bombardeos_51_52['NBR_ABORT_AIRCRAFT'].fillna(0)
df_bombardeos_51_52['NBR_ATTACK_EFFEC_AIRCRAFT'] = df_bombardeos_51_52['NBR_ATTACK_EFFEC_AIRCRAFT'].fillna(0)
mission_abort_rate = df_bombardeos_51_52['NBR_ABORT_AIRCRAFT'].sum() / total_effective_aircraft
print(f"Tasa de abortos de misi√≥n: {mission_abort_rate:.2%}")

# P√©rdidas de aeronaves por misi√≥n
# Primero, convertimos la columna 'NBR_LOST_AIRCRAFT' a un tipo num√©rico, ignorando errores
df_bombardeos_51_52['NBR_LOST_AIRCRAFT'] = pd.to_numeric(df_bombardeos_51_52['NBR_LOST_AIRCRAFT'], errors='coerce')
average_aircraft_losses = df_bombardeos_51_52['NBR_LOST_AIRCRAFT'].mean()
print(f"P√©rdidas promedio de aeronaves por misi√≥n: {average_aircraft_losses:.2f}")

# Tipos de misi√≥n con mayores p√©rdidas
losses_by_mission_type = df_bombardeos_51_52.groupby('MISSION_TYPE')['NBR_LOST_AIRCRAFT'].mean().sort_values(ascending=False)
print("\nTipos de misi√≥n con mayores p√©rdidas de aeronaves:")
print(losses_by_mission_type)

# Eficiencia de las misiones
df_bombardeos_51_52['efficiency'] = df_bombardeos_51_52['NBR_ATTACK_EFFEC_AIRCRAFT'] / (df_bombardeos_51_52['NBR_ATTACK_EFFEC_AIRCRAFT'] + df_bombardeos_51_52['NBR_ABORT_AIRCRAFT'])
average_efficiency = df_bombardeos_51_52['efficiency'].mean()
print(f"\nEficiencia promedio de las misiones: {average_efficiency:.2%}")
```

```python
# Frecuencia de tipos de objetivos
target_type_frequency = df_bombardeos_51_52['TGT_TYPE'].value_counts()
print("Frecuencia de tipos de objetivos:")
print(target_type_frequency)

# Distribuci√≥n geogr√°fica de objetivos
plt.figure(figsize=(10, 6))
plt.scatter(df_bombardeos_51_52['TGT_LONGITUDE_WGS84'], df_bombardeos_51_52['TGT_LATITUDE_WGS84'], alpha=0.5, c='red', s=10)
plt.title('Distribuci√≥n geogr√°fica de objetivos')
plt.xlabel('Longitud')
plt.ylabel('Latitud')
plt.grid(True)
plt.show()

# Eficiencia por tipo de objetivo
efficiency_by_target_type = df_bombardeos_51_52.groupby('TGT_TYPE')['efficiency'].mean()
print("\nEficiencia por tipo de objetivo:")
print(efficiency_by_target_type)
```

```python
# Promedio de armas utilizadas por misi√≥n
df_bombardeos_51_52['NBR_OF_WEAPONS'] = pd.to_numeric(df_bombardeos_51_52['NBR_OF_WEAPONS'], errors='coerce')
average_weapons_per_mission = df_bombardeos_51_52['NBR_OF_WEAPONS'].mean()
print(f"Promedio de armas utilizadas por misi√≥n: {average_weapons_per_mission:.2f}")

# Carga promedio de bombas por misi√≥n
average_bombload_per_mission = df_bombardeos_51_52['TOTAL_BOMBLOAD_IN_LBS'].mean()
print(f"Carga promedio de bombas por misi√≥n: {average_bombload_per_mission:.2f}")

# Uso de tipos de armas
weapons_type_frequency = df_bombardeos_51_52['WEAPONS_TYPE'].value_counts()
print("Frecuencia de tipos de armas:")
print(weapons_type_frequency)
```

```python
# Agrupar por mes y calcular m√©tricas clave
monthly_metrics = df_bombardeos_51_52.groupby('MONTH').agg({
    'efficiency': 'mean',
    'NBR_LOST_AIRCRAFT': 'sum',
    'NBR_ABORT_AIRCRAFT': 'sum',
    'MISSION_NUMBER': 'count'
}).reset_index()

# Graficar misiones y eficiencia promedio por mes
fig, ax1 = plt.subplots(figsize=(14, 7))

ax1.set_xlabel('Mes')
ax1.set_ylabel('N√∫mero de misiones', color='tab:blue')
ax1.plot(monthly_metrics['MONTH'], monthly_metrics['MISSION_NUMBER'], marker='o', linestyle='-', color='tab:blue', label='Misiones por mes')
ax1.tick_params(axis='y', labelcolor='tab:blue')

ax2 = ax1.twinx()
ax2.set_ylabel('Eficiencia promedio', color='tab:green')
ax2.plot(monthly_metrics['MONTH'], monthly_metrics['efficiency'], marker='o', linestyle='--', color='tab:green', label='Eficiencia promedio')
ax2.tick_params(axis='y', labelcolor='tab:green')

fig.tight_layout()
fig.legend(loc='upper left', bbox_to_anchor=(0.1, 0.9))
plt.title('Tendencia y eficiencia promedio por mes')
plt.show()

# P√©rdidas y abortos por mes
plt.figure(figsize=(14, 7))
plt.plot(monthly_metrics['MONTH'], monthly_metrics['NBR_LOST_AIRCRAFT'], marker='o', linestyle='--', color='tab:red', label='P√©rdidas de aeronaves')
plt.plot(monthly_metrics['MONTH'], monthly_metrics['NBR_ABORT_AIRCRAFT'], marker='o', linestyle=':', color='tab:orange', label='Abortos de misi√≥n')
plt.title('P√©rdidas y abortos de aeronaves por mes')
plt.xlabel('Mes')
plt.ylabel('N√∫mero de aeronaves')
plt.legend()
plt.grid(True)
plt.show()
```

```python
# Tasa de da√±os en objetivos
bda_success_rate = df_bombardeos_51_52['BDA'].value_counts(normalize=True)
print("Tasa de da√±os en objetivos (BDA):")
print(bda_success_rate)

# Relaci√≥n entre carga de bombas y √©xito
df_bombardeos_51_52['TOTAL_BOMBLOAD_IN_LBS'] = pd.to_numeric(df_bombardeos_51_52['TOTAL_BOMBLOAD_IN_LBS'], errors='coerce')
df_bombardeos_51_52['BDA_numeric'] = pd.to_numeric(df_bombardeos_51_52['BDA'], errors='coerce')
bombload_success_correlation = df_bombardeos_51_52[['TOTAL_BOMBLOAD_IN_LBS', 'BDA_numeric']].dropna().corr()
print("\nCorrelaci√≥n entre carga de bombas y √©xito (BDA):")
print(bombload_success_correlation)

# Relaci√≥n entre altura de vuelo y efectividad
altitude_success_correlation = df_bombardeos_51_52[['ALTITUDE_FT', 'BDA_numeric']].dropna().corr()
print("\nCorrelaci√≥n entre altura de vuelo y √©xito (BDA):")
print(altitude_success_correlation)
```

```python
# Agrupar por 'City/Town' y contar los objetivos m√°s comunes
common_targets_by_city = df_bombardeos_51_52['City/Town'].value_counts()
print("Ubicaci√≥n de objetivos m√°s frecuentes:")
print(common_targets_by_city.head(10))

# Visualizar las ubicaciones de lanzamiento y los objetivos en mapas de calor
import folium
from folium.plugins import HeatMap

# Convertir coordenadas a valores num√©ricos
def convert_to_float(coord):
    try:
        return float(coord)
    except ValueError:
        return None

df_bombardeos_51_52['SOURCE_TGT_LAT'] = df_bombardeos_51_52['SOURCE_TGT_LAT'].apply(convert_to_float)
df_bombardeos_51_52['SOURCE_TGT_LONG'] = df_bombardeos_51_52['SOURCE_TGT_LONG'].apply(convert_to_float)

# Crear un mapa base
mapa = folium.Map(location=[38.5, 127], zoom_start=6)

# Agregar ubicaciones de lanzamiento como un mapa de calor
heat_data_launch = [[row['SOURCE_TGT_LAT'], row['SOURCE_TGT_LONG']] for index, row in df_bombardeos_51_52.dropna(subset=['SOURCE_TGT_LAT', 'SOURCE_TGT_LONG']).iterrows()]
HeatMap(heat_data_launch, radius=10, max_zoom=13).add_to(mapa)

# Agregar ubicaciones de objetivos como un mapa de calor
heat_data_target = [[row['TGT_LATITUDE_WGS84'], row['TGT_LONGITUDE_WGS84']] for index, row in df_bombardeos_51_52.dropna(subset=['TGT_LATITUDE_WGS84', 'TGT_LONGITUDE_WGS84']).iterrows()]
HeatMap(heat_data_target, radius=10, max_zoom=13).add_to(mapa)

# Mostrar el mapa
mapa

# Analizar misiones terrestres vs mar√≠timas
land_missions = df_bombardeos_51_52[df_bombardeos_51_52['In_Land'] == True]
water_missions = df_bombardeos_51_52[df_bombardeos_51_52['In_Land'] == False]

# Comparar m√©tricas clave
land_missions_metrics = land_missions.agg({
    'efficiency': 'mean',
    'NBR_LOST_AIRCRAFT': 'sum',
    'NBR_ABORT_AIRCRAFT': 'sum',
    'MISSION_NUMBER': 'count'
})

water_missions_metrics = water_missions.agg({
    'efficiency': 'mean',
    'NBR_LOST_AIRCRAFT': 'sum',
    'NBR_ABORT_AIRCRAFT': 'sum',
    'MISSION_NUMBER': 'count'
})

print("\nM√©tricas de misiones terrestres:")
print(land_missions_metrics)

print("\nM√©tricas de misiones mar√≠timas:")
print(water_missions_metrics)
```

```python
# Verificar consistencia entre CALCULATED_BOMBLOAD_LBS y TOTAL_BOMBLOAD_IN_LBS
df_bombardeos_51_52['CALCULATED_BOMBLOAD_LBS'] = pd.to_numeric(df_bombardeos_51_52['CALCULATED_BOMBLOAD_LBS'], errors='coerce')
df_bombardeos_51_52['TOTAL_BOMBLOAD_IN_LBS'] = pd.to_numeric(df_bombardeos_51_52['TOTAL_BOMBLOAD_IN_LBS'], errors='coerce')
bombload_consistency = df_bombardeos_51_52[['CALCULATED_BOMBLOAD_LBS', 'TOTAL_BOMBLOAD_IN_LBS']].dropna()
consistency_check = bombload_consistency['CALCULATED_BOMBLOAD_LBS'] == bombload_consistency['TOTAL_BOMBLOAD_IN_LBS']
consistency_rate = consistency_check.mean()
print(f"Consistencia entre CALCULATED_BOMBLOAD_LBS y TOTAL_BOMBLOAD_IN_LBS: {consistency_rate:.2%}")

# Comparar m√©todos de avistamiento con el √©xito medido por BDA
bda_success_by_sighting_method = df_bombardeos_51_52.groupby('BOMB_SIGHTING_METHOD')['BDA_numeric'].mean()
print("\n√âxito medido por BDA seg√∫n el m√©todo de avistamiento:")
print(bda_success_by_sighting_method)

# Correlaciones clave
# NBR_OF_WEAPONS vs. BDA
weapons_bda_correlation = df_bombardeos_51_52[['NBR_OF_WEAPONS', 'BDA_numeric']].dropna().corr()
print("\nCorrelaci√≥n entre NBR_OF_WEAPONS y BDA:")
print(weapons_bda_correlation)

# ALTITUDE_FT vs. NBR_LOST_AIRCRAFT
altitude_losses_correlation = df_bombardeos_51_52[['ALTITUDE_FT', 'NBR_LOST_AIRCRAFT']].dropna().corr()
print("\nCorrelaci√≥n entre ALTITUDE_FT y NBR_LOST_AIRCRAFT:")
print(altitude_losses_correlation)
```
